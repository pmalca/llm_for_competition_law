```python
import openai
from openai import OpenAI
client = OpenAI(api_key="YOUR_API_KEY")
import pandas as pd
```

```python
df_2009 = pd.read_excel("/clc_2009.xlsx")
df_ft = pd.read_excel("/data_for_ft.xlsx")
```

```python
# We need to remove the the rows that df_2009 and df_ft share, so we get a dataset with no classification

enlace_values_ft = df_ft['Enlace'].unique()

# Step 2: Filter df_2009 to keep rows where 'Enlace' is not in enlace_values_ft
final_df = df_2009[~df_2009['Enlace'].isin(enlace_values_ft)]
final_df
```

```python
def generate_output(text):
    try:
        # Create chat completion
        completion = client.chat.completions.create(
            model="ft:gpt-3.5-turbo-0125:personal:competition-law:9Xy3cuWb",
            messages=[
                {"role": "system", "content": "Vas a clasificar un texto legal. Debes determinar si hubo una intervención indicando 'Si' o 'No'. Luego, debes identificar si corresponde a la materia 'concentraciones', 'conductas' u 'otros'. Luego, debes clasificar si en la categoría de concentraciones hay descripción de 'sin condiciones' o 'condiciones'. Finalmente, debes clasificar si en la descripción de conductas se trata de 'cooperativa' o 'abusiva'."},
                {"role": "user", "content": text}
            ]
        )
        return completion.choices[0].message.content
    except Exception as e:
        print(f"An error occurred: {e}")
        # Code to execute in case of error goes here

final_df['Clasificación'] = final_df['Sumilla'].apply(generate_output)
# final_df.to_csv(r"\clc_2009_c.csv")
# final_df.to_excel(r"\clc_2009_c.xlsx")
```

```python
competition_df = pd.concat([df_ft, final_df], ignore_index=True)
competition_df
```

```python
def limpiar_texto(texto):
    texto_limpio = texto.replace('\n', ' ')
    return texto_limpio
competition_df['Clasificación'] =competition_df['Clasificación'].apply(limpiar_texto)
```

```python
# we will use regex to finish the classification task. 

import re
# Define patterns
pattern1 = r'intervención:'
pattern2 = r'materia'

# Extract text using regex
def extract_text(row):
    match1 = re.search(pattern1, row['Clasificación'], flags=re.IGNORECASE)
    match2 = re.search(pattern2, row['Clasificación'], flags=re.IGNORECASE)
    if match1:
        start_index = match1.end()
    else:
        return ""

    if match2:
        end_index = match2.start()
    else:
        end_index = len(row['Clasificación'])

    return row['Clasificación'][start_index:end_index]

competition_df['Intervención'] = competition_df.apply(extract_text, axis=1)
```

```python
# Define patterns
pattern1 = r'materia:'
pattern2 = r'descripción'

# Extract text using regex
def extract_text(row):
    match1 = re.search(pattern1, row['Clasificación'], flags=re.IGNORECASE)
    match2 = re.search(pattern2, row['Clasificación'], flags=re.IGNORECASE)
    if match1:
        start_index = match1.end()
    else:
        return ""

    if match2:
        end_index = match2.start()
    else:
        end_index = len(row['Clasificación'])

    return row['Clasificación'][start_index:end_index]

competition_df['Materia'] = competition_df.apply(extract_text, axis=1)
```

```python
def extract_description(text):
    match = re.search(r'descripción:\s*(.*)', text, re.IGNORECASE)
    if match:
        return match.group(1)
    else:
        return None

# Apply the function to the 'Text' column
competition_df['Descripción'] = competition_df['Clasificación'].apply(extract_description)
```
